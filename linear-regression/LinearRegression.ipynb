```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_sedf.delection import train_test_split
```

```python
df = pd.read_csv("ecommerce.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.jointplot(x="Time on Website", y="Yearly Amount Spent", data=df, alpha=0.5)
```

```python
sns.jointplot(x="Time on App", y="Yearly Amount Spent", data=df, alpha=0.5)
```

```python
sns.pairplot(df, kind='scatter', plot_kws={'alpha':0.4})
```

```python
X = df[['Time on App','Time on Website', 'Avg. Session Length', 'Length of Membership']]
y = df['Yearly Amount Spent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=6)
# 0.3 -> 30% of the set dedicated to testing, and 70% for training
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
y_train
```

```python
# now actually training the model: making model
linearmodel = LinearRegression()
```

```python
linearmodel.fit(X_train, y_train)
```

```python
linearmodel.coef_
```

```python
cdf = pd.DataFrame(linearmodel.coef_, X.columns, columns = ['Coef'])
print(cdf)
```

```python
### in the above boxes, the different weights are under Coef, where the larger weights contribute more to the predicted value
```

```python
# predictions for data
pred = linearmodel.predict(X_test)
```

```python
pred
```

```python
sns.scatterplot(x=pred, y=y_test)
plt.xlabel("Predicted Values")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
# need math for sqrt func
```

```python
print("Mean Absolute Error:", mean_absolute_error(y_test, pred))
print("Mean Squared Error:", mean_squared_error(y_test, pred))
print("Root Mean Squared Error:", math.sqrt(mean_squared_error(y_test, pred)))
```

```python
# residual analysis
residual = y_test - pred
```

```python
sns.displot(residual, bins=30, kde=True)
```

```python
import pylab
import scipy.stats as stats

stats.probplot(residual, dist="norm", plot=pylab)
pylab.show()
```

```python
from sklearn.metrics import r2_score
print("R^2 Score:", r2_score(y_test, pred))
# R^2 score is a measure of how well the model explains the variance in the data
# R^2 score close to 1 indicates a good fit, while close to 0 indicates a poor fit
```

```python
## Above shows a 98% accuracy level
# Making a prediction for a new customer
# 'Time on App','Time on Website', 'Avg. Session Length', 'Length of Membership'
new_customer = pd.DataFrame({
    'Time on App': [12.5],
    'Time on Website': [39.0],
    'Avg. Session Length': [34.5],
    'Length of Membership': [4.0]
})
new_prediction = linearmodel.predict(new_customer)
print("Predicted Yearly Amount Spent for new customer:", new_prediction[0])
# Visualizing the relationship between features and target variable
sns.pairplot(df, x_vars=['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership'], y_vars='Yearly Amount Spent', height=5, aspect=0.7)
plt.show()
```

```python
# saving the model
import joblib
joblib.dump(linearmodel, 'ecommerce_model.pkl')
```
